分布式训练结果

```python
res_ada_r0 = np.load('dist/resnet数值ada64rank0#.npz')
res_ada_r1 = np.load('dist/resnet数值ada64rank1#.npz')
res_ada_r2 = np.load("dist/resnet数值ada64rank2#.npz")

vgg_ada_r0 = np.load('dist/vgg19_bn数值ada64rank0#.npz')
vgg_ada_r1 = np.load('dist/vgg19_bn数值ada64rank1#.npz')
vgg_ada_r2 = np.load('dist/vgg19_bn数值ada64rank2#.npz')
```

```python
res_ada_r2.files
```

```python
res_fp_mean = (res_ada_r0['FP_time'] + res_ada_r1['FP_time'] + res_ada_r2['FP_time']) / 3
res_bp_mean = (res_ada_r0['BP_time'] + res_ada_r1['BP_time'] + res_ada_r2['BP_time']) / 3

vgg_fp_mean = (vgg_ada_r0['FP_time'] + vgg_ada_r1['FP_time'] + vgg_ada_r2['FP_time']) / 3
vgg_bp_mean = (vgg_ada_r0['BP_time'] + vgg_ada_r1['BP_time'] + vgg_ada_r2['BP_time']) / 3
```

```python
res_multi_ada = [res_fp_mean, res_bp_mean]  # (447.70, 842.14)
res_ada = [998.05, 1993.58]
res_fix128 = [1024.36, 2105.98]
```

```python
vgg_multi_ada = [vgg_fp_mean, vgg_bp_mean]  # (733.82, 3399.22)
vgg_ada = [2342.10, 4865.32]
vgg_fix128 = [2406.40, 5160.00]
```

```python
range(3), np.arange(3)
```

```python
labels = ['fix128', 'ada128-2048', 'multi_ada']

X = np.arange(3)
fp_time = [res_fix128[0], res_ada[0], res_multi_ada[0]]
bp_time = [res_fix128[1], res_ada[1], res_multi_ada[1]]

fp_speedup = [1.0, res_fix128[0]/res_ada[0], res_fix128[0]/res_multi_ada[0]]
bp_speedup = [1.0, res_fix128[1]/res_ada[1], res_fix128[1]/res_multi_ada[1]]

fig_, ax3 = plt.subplots()
ax4 = ax3.twinx()  # 做镜像处理，好画两个刻度不同的 y 轴

ax3.bar(X, fp_time, width=0.3, tick_label=labels, label='fp time')
ax3.bar(X+0.3, bp_time, width=0.3, label='bp time')
ax3.set_ylabel('time /s')
ax3.set_xlabel('batch schedular \n\n ResNet-20')

# fp_speedup 折线
ax4.plot(X, fp_speedup, marker='s', color='r', label='fp_speedup')
for x, y in zip(X, fp_speedup):
    ax4.text(x, y, '%.2f' % y, fontdict={'fontsize':11})  # 显示点的数值

# bp_speedup 折线
ax4.plot(X+0.3, bp_speedup,  marker='v', color='k', label='bp_speedup')
for x, y in zip(X+0.3, bp_speedup):
    ax4.text(x, y, '%.2f' % y, fontdict={'fontsize':11})

ax4.set_ylabel('speedup')
ax4.legend(loc='upper center')  # 图例位置

ax3.set_ylim(0, 2700)
ax3.legend(loc='center right')
ax3.grid()

plt.show()
```

```python
labels = ['fix128', 'ada128-2048', 'multi_ada']

X = np.arange(3)
fp_time = [vgg_fix128[0], vgg_ada[0], vgg_multi_ada[0]]
bp_time = [vgg_fix128[1], vgg_ada[1], vgg_multi_ada[1]]

fp_speedup = [1.0, vgg_fix128[0]/vgg_ada[0], vgg_fix128[0]/vgg_multi_ada[0]]
bp_speedup = [1.0, vgg_fix128[1]/vgg_ada[1], vgg_fix128[1]/vgg_multi_ada[1]]

fig_, ax3 = plt.subplots()
ax4 = ax3.twinx()  # 做镜像处理，好画两个刻度不同的 y 轴

ax3.bar(X, fp_time, width=0.3, tick_label=labels, label='fp time')
ax3.bar(X+0.3, bp_time, width=0.3, label='bp time')
ax3.set_ylabel('time /s')
ax3.set_xlabel('batch schedular \n\n VGG19_BN')

# fp_speedup 折线
ax4.plot(X, fp_speedup, marker='s', color='r', label='fp_speedup')
for x, y in zip(X, fp_speedup):
    ax4.text(x, y, '%.2f' % y, fontdict={'fontsize':11}, 
             horizontalalignment='center', verticalalignment='bottom')  # 显示点的数值

# bp_speedup 折线
ax4.plot(X+0.3, bp_speedup,  marker='v', color='k', label='bp_speedup')
for x, y in zip(X+0.3, bp_speedup):
    ax4.text(x, y, '%.2f' % y, fontdict={'fontsize':11}, 
             horizontalalignment='center', verticalalignment='bottom')  # 'top', 'bottom', 'center', 'baseline', 'center_baseline'

ax4.set_ylabel('speedup')
ax4.legend(loc='upper center')  # 图例位置

ax3.set_ylim(0, 6500)
ax4.set_ylim(0.8, 3.6)
ax3.legend(bbox_to_anchor=(1., 0.8))
ax3.grid()

plt.show()
```
