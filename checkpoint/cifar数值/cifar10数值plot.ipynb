```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alex_ada = np.load('alexnet数值ada128#cifar10.npz')
alex_fix = np.load('alexnet数值fixed256#cifar10.npz')
alex_fix2 = np.load('alexnet数值fixed4096_cifar10.npz')

resnet_ada = np.load('resnet数值ada64#cifar10.npz')
resnet_fix = np.load('resnet数值fixed128#cifar10.npz')
resnet_fix2 = np.load('resnet数值fixed2048_cifar10.npz')

vgg_ada = np.load('vgg19_bn数值ada64#cifar10.npz')
vgg_fix = np.load('vgg19_bn数值fixed128#cifar10.npz')
vgg_fix2 = np.load('vgg19_bn数值fixed2048_cifar10.npz')


epoch = 100
```

```python
alex_ada_loss, alex_ada_acc = alex_ada['loss'], alex_ada['acc']
alex_fix_loss, alex_fix_acc = alex_fix['loss'], alex_fix['acc']
alex_fix2_loss, alex_fix2_acc = alex_fix2['loss'], alex_fix2['acc']

resnet_ada_loss, resnet_ada_acc = resnet_ada['loss'], resnet_ada['acc']
resnet_fix_loss, resnet_fix_acc = resnet_fix['loss'], resnet_fix['acc']
resnet_fix2_loss, resnet_fix2_acc = resnet_fix2['loss'], resnet_fix2['acc']

vgg_ada_loss, vgg_ada_acc = vgg_ada['loss'], vgg_ada['acc']
vgg_fix_loss, vgg_fix_acc = vgg_fix['loss'], vgg_fix['acc']
vgg_fix2_loss, vgg_fix2_acc = vgg_fix2['loss'], vgg_fix2['acc']
```

```python
ffig, (fax1, fax2, fax3) = plt.subplots(1, 3, figsize=(12, 4))
ffig.subplots_adjust(wspace=0.3)

fax1.plot(range(1, epoch+1), vgg_ada_loss, label='ada128-2048')
fax1.plot(range(1, epoch+1), vgg_fix_loss, '--', label='fix128')
fax1.plot(range(1, epoch+1), vgg_fix2_loss, label='fix2048')
fax1.set_xlabel('epochs \n\n (a) VGG19_BN')
fax1.set_ylabel('loss')
fax1.legend()
fax1.grid()

fax2.plot(range(1, epoch+1), resnet_ada_loss, label='ada128-2048')
fax2.plot(range(1, epoch+1), resnet_fix_loss, '--', label='fix128')
fax2.plot(range(1, epoch+1), resnet_fix2_loss, label='fix2048')
fax2.set_xlabel('epochs \n\n (b) ResNet-20')
fax2.set_ylabel('loss')
fax2.legend()
fax2.grid()

fax3.plot(range(1, epoch+1), alex_ada_loss, label='ada256-4096')
fax3.plot(range(1, epoch+1), alex_fix_loss, '--', label='fix256')
fax3.plot(range(1, epoch+1), alex_fix2_loss, label='fix4096')
fax3.set_xlabel('epochs \n\n (c) AlexNet')
fax3.set_ylabel('loss')
fax3.legend()
fax3.grid()

plt.show()
```

```python
ffig, (fax1, fax2, fax3) = plt.subplots(1, 3, figsize=(12, 4))
ffig.subplots_adjust(wspace=0.3)

fax1.plot(range(1, epoch+1), [100 - x for x in vgg_ada_acc], label='ada128-2048')
fax1.plot(range(1, epoch+1), [100 - x for x in vgg_fix_acc], '--', label='fix128')
fax1.plot(range(1, epoch+1), [100 - x for x in vgg_fix2_acc], label='fix2048')
fax1.set_xlabel('epochs \n\n (a) VGG19_BN')
fax1.set_ylabel('test error')
fax1.legend()
fax1.grid()

fax2.plot(range(1, epoch+1), [100 - x for x in resnet_ada_acc], label='ada128-2048')
fax2.plot(range(1, epoch+1), [100 - x for x in resnet_fix_acc], '--', label='fix128')
fax2.plot(range(1, epoch+1), [100 - x for x in resnet_fix2_acc], label='fix2048')
fax2.set_xlabel('epochs \n\n (b) ResNet-20')
fax2.set_ylabel('test error')
fax2.legend()
fax2.grid()

fax3.plot(range(1, epoch+1), [100 - x for x in alex_ada_acc], label='ada256-4096')
fax3.plot(range(1, epoch+1), [100 - x for x in alex_fix_acc], '--', label='fix256')
fax3.plot(range(1, epoch+1), [100 - x for x in alex_fix2_acc], label='fix4096')
fax3.set_xlabel('epochs \n\n (c) AlexNet')
fax3.set_ylabel('test error')
fax3.legend()
fax3.grid()

plt.show()
```
